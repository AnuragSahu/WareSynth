```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from os import listdir
from os.path import isfile, join
```

```python
dataset_path = "../../../test_images/static/"


files = [f for f in listdir(dataset_path) if isfile(join(dataset_path, f))]
files.sort()

# to find the number of boxes in the scene
for file in files:
    img = cv2.imread(join(dataset_path, file), 0)
    img = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)[1]  # ensure binary
    num_labels, labels_im = cv2.connectedComponents(img)
    number_of_boxes = num_labels - 1
    print(file, " : ", number_of_boxes)
```

```python
def getBoundingBoxes(img):
    boundingBoxes = []
    contours = cv2.findContours(img, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    contours = contours[0] if len(contours) == 2 else contours[1]
    for cntr in contours:
        x,y,w,h = cv2.boundingRect(cntr)
        boundingBoxes.append([x,y,w,h])
    return boundingBoxes

def rackAndBoxBBs(img):
    gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    boxThresh = cv2.threshold(gray,128,255,cv2.THRESH_BINARY)[1]
    boxBB = getBoundingBoxes(boxThresh)

    shelfThresh = cv2.threshold(gray,10,155,cv2.THRESH_BINARY)[1]
    rackBB = getBoundingBoxes(shelfThresh)
    
    return rackBB, boxBB

def scrutiniseFreeSpace(freeSpaceBBox):
    
```
